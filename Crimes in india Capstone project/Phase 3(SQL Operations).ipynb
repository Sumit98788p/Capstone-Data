3.8.2	Write SQL query to find which District in each state/ut has the highest number of murders yearwise. Your output should show STATE/UT, YEAR, DISTRICT, and MURDERS.





```python
Select c.state,c.district,c.year,c.murder from crimescommitedipc c join
(select max(murder) as Max_murder,district from crimescommitedipc group by district) r 
on c.murder = r.Max_murder and c.district = r.district
```

3.8.3	Store the above data (the result of 3.8.2) in DataFrame and analyze districts that appear 3 or more than 3 years and print the corresponding state/ut, district, murders, and year in descending order.


```python
df = pd.read_csv("https://raw.githubusercontent.com/Sumit98788p/Internship/main/Exported%20data%20of%203.8.2.csv")
df
```

```python
df['district'].duplicated()
```

```python
df.loc[df['district'].duplicated(),:]
```

```python
district_counts = df.groupby(['state', 'district']).size()

# Filter districts appearing 3 or more years
districts_appearing_multiple_years = district_counts[district_counts >= 3].reset_index()

# Merge with original data to get all details
merged_df = pd.merge(districts_appearing_multiple_years, df, how='left', on=['state', 'district'])

# Print the result in descending order of year
result = merged_df[['state', 'district', 'murder', 'year']].sort_values(by='year', ascending=False)
print("Districts Appearing 3 or More Years with Corresponding State/UT, District, Murders, and Year (Descending Order):")
print(result)
```
